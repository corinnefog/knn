````python
from tensorflow.keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test


import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
````

````python
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

````python
X_train_flattened = X_train.reshape(X_train.shape[0], 28*28) 
X_test_flattened = X_test.reshape(X_test.shape[0], 28*28)  
````

````python
from collections import Counter

def euclidean_distance(x1, x2):
    """Compute Euclidean distance between two vectors."""
    return np.sqrt(np.sum((x1 - x2) ** 2))

def predict_knn(X_train, y_train, X_test, k=3):
    """Predict labels using k-Nearest Neighbors."""
    y_pred = []
    
    for test_sample in X_test:
        # Compute distance to all training points
        distances = [euclidean_distance(test_sample, train_sample) for train_sample in X_train]
        
        # Get k nearest neighbors (smallest distances)
        k_indices = np.argsort(distances)[:k]
        k_nearest_labels = [y_train[i] for i in k_indices]
        
        # Majority vote for the final prediction
        most_common = Counter(k_nearest_labels).most_common(1)[0][0]
        y_pred.append(most_common)
    
    return np.array(y_pred)

def compute_accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

# Try different k values
for k in [1, 3, 5, 7, 9]:
    y_pred = predict_knn(X_train_flattened[:1000], y_train[:1000], X_test_flattened[:100], k)  # Limit samples for speed
    acc = compute_accuracy(y_test[:100], y_pred)
    print(f"Accuracy with k={k}: {acc:.4f}")
````

````python
def compute_accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)

accuracy = compute_accuracy(y_test[:100], y_pred)
print(f"Accuracy: {accuracy:.4f}")

def confusion_matrix(y_true, y_pred, num_classes=10):
    matrix = np.zeros((num_classes, num_classes), dtype=int)
    
    for true, pred in zip(y_true, y_pred):
        matrix[true, pred] += 1  # Count occurrences
    
    return matrix

cm = confusion_matrix(y_test[:100], y_pred)
print("Confusion Matrix:\n", cm)
````
